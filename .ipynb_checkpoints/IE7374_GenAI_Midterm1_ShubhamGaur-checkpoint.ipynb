### 1.3. Data Preprocessing
Text data is cleaned by removing special characters and converting to lowercase.

```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def clean_text(text):
    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)  # Remove URLs
    text = re.sub(r"[^a-zA-Z\s]", "", text)  # Remove special characters
    words = [lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words]
    return " ".join(words)

df["clean_text"] = df["text"].apply(clean_text)

print(df[["text", "clean_text"]].head())
```

The dataset undergoes preprocessing to remove inconsistencies and prepare the text for model
input. The key cleaning steps include:

__Handling missing values__ – Checking for null or empty fields in the dataset and removing or
imputing missing entries.

__Removing duplicate entries__ – Ensuring the dataset does not contain redundant data points
that could bias training.

__Standardizing text__ – Converting all text to lowercase to maintain consistency.

__Removing special characters and unnecessary whitespace__ – Keeping only alphanumeric
characters and punctuation relevant to semantic meaning.

__Tokenization__ – Converting textual data into a format that can be processed by the tokenizer
of the selected LLMs.

### Term Frequency Analysis
is conducted to identify key terms that strongly correlate with
certain emotions.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
from nltk.tokenize import word_tokenize
import nltk

# Download necessary tokenizer
nltk.download('punkt')

# Function to compute term frequencies for a given text dataset
def get_term_frequencies(text_series, top_n=10):
    all_words = " ".join(text_series).split()  # Combine all text and split into individual words
    word_counts = Counter(all_words)  # Count occurrences of each unique word
    return word_counts.most_common(top_n)  # Retrieve the top N most frequent words

# Extract unique emotion categories from the dataset
emotions = df["emotion"].unique()

# Dictionary to store term frequency data for each emotion
emotion_term_frequencies = {}

# Compute term frequency for each emotion class
for emotion in emotions:
    emotion_text = df[df["emotion"] == emotion]["clean_text"]  # Filter dataset for specific emotion
    emotion_term_frequencies[emotion] = get_term_frequencies(emotion_text, top_n=10)

# Convert frequency data into a structured DataFrame for visualization
term_freq_df = pd.DataFrame([
    {"Emotion": emotion, "Word": word, "Frequency": freq}
    for emotion, terms in emotion_term_frequencies.items()
    for word, freq in terms
])

# Plot term frequency distribution for each emotion
plt.figure(figsize=(12, 6))
sns.barplot(data=term_freq_df, x="Word", y="Frequency", hue="Emotion", dodge=True, palette="viridis")

# Format the plot
plt.xticks(rotation=45)
plt.xlabel("Words", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Top 10 Frequent Words per Emotion", fontsize=14)
plt.legend(title="Emotion")

# Display the plot
plt.show()
```

The term frequency analysis reveals the most common words associated with each
emotion. The word "friend" appears frequently across multiple emotions, indicating its
strong impact on emotional experiences. Words like "happy" and "exam" are dominant in
joy, while "afraid", "alone", and "night" characterize fear. Anger is often linked to "angry",
"mother", and "time", reflecting interpersonal conflicts. Sadness is associated with "died",
"father", and "death", emphasizing loss. Disgust includes terms like "disgusted",
"people", and "mean", highlighting unpleasant interactions. Shame and guilt are driven by
words such as "ashamed", "something", and "guilty", reflecting personal regret. This
analysis helps in understanding how words strongly correlate with specific emotions,
enhancing text-based emotion classification.

### 1.4.4. Inter-Class Similarity

To examine the overlap between different emotion labels:

### t-SNE Plot 

is generated to visualize the distribution of text embeddings in lower-dimensional
space.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from transformers import AutoTokenizer, AutoModel
import torch
from tqdm import tqdm  # For progress tracking

# Load a pre-trained transformer model for text embedding
model_name = "bert-base-uncased"  # Change if needed (e.g., "distilbert-base-uncased")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Function to generate embeddings using the CLS token representation
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():  # Disable gradient computation for efficiency
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].squeeze().numpy()  # Extract CLS token embedding

# Compute embeddings for the entire dataset
embeddings = []
for text in tqdm(df["clean_text"], desc="Extracting Embeddings"):
    embeddings.append(get_embedding(text))

# Convert embeddings to a NumPy array
embeddings = np.array(embeddings)

# Standardize embeddings to ensure uniform scaling
scaler = StandardScaler()
embeddings_scaled = scaler.fit_transform(embeddings)

# Apply PCA to reduce dimensions before t-SNE
pca = PCA(n_components=50)  # Reduce embeddings to 50 dimensions
embeddings_pca = pca.fit_transform(embeddings_scaled)

# Apply t-SNE for visualization in 2D space
tsne = TSNE(n_components=2, perplexity=50, random_state=42, n_iter=5000)
embeddings_2d = tsne.fit_transform(embeddings_pca)

# Convert to DataFrame for visualization
df_tsne = pd.DataFrame(embeddings_2d, columns=["Dim1", "Dim2"])
df_tsne["Emotion"] = df["emotion"].values  # Assign emotion labels for plotting

# Plot t-SNE visualization
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df_tsne, x="Dim1", y="Dim2", hue="Emotion", palette="tab10", alpha=0.8)

# Formatting the plot
plt.title("t-SNE Visualization of Text Embeddings with PCA")
plt.xlabel("t-SNE Dimension 1")
plt.ylabel("t-SNE Dimension 2")
plt.legend(title="Emotion", bbox_to_anchor=(1.05, 1), loc="upper left")

# Display the plot
plt.show()
```

The t-SNE visualization of text embeddings shows the clustering of different emotions in a
lower-dimensional space. While some emotions exhibit clear separation, others have
overlapping regions, indicating semantic similarities in emotional expressions. The presence
of a distinct cluster suggests a subset of text samples with unique characteristics. This
visualization highlights the challenge of emotion classification, as certain emotions share
linguistic patterns, making them harder to differentiate. Fine-tuning the model and using
better embedding techniques may improve separation.

### Cosine Similarity 
between text embeddings of different emotion classes is computed to
identify closely related emotional expressions.

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity
from transformers import AutoTokenizer, AutoModel
import torch
from tqdm import tqdm  # For progress tracking

# Load a pre-trained transformer model for text embeddings
model_name = "bert-base-uncased"  # Change if needed (e.g., "distilbert-base-uncased")
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModel.from_pretrained(model_name)

# Function to generate sentence embeddings using CLS token representation
def get_embedding(text):
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=128)
    with torch.no_grad():  # Disable gradient computation for efficiency
        outputs = model(**inputs)
    return outputs.last_hidden_state[:, 0, :].squeeze().numpy()  # Extract CLS token embedding

# Compute average embeddings for each emotion class
emotion_embeddings = {}
for emotion in df["emotion"].unique():
    texts = df[df["emotion"] == emotion]["clean_text"].sample(n=500, random_state=42)  # Sample 500 texts per class
    embeddings = np.array([get_embedding(text) for text in tqdm(texts, desc=f"Processing {emotion}")])
    emotion_embeddings[emotion] = np.mean(embeddings, axis=0)  # Compute mean embedding for each emotion class

# Convert emotion embeddings into a matrix format
emotion_labels = list(emotion_embeddings.keys())
embedding_matrix = np.array([emotion_embeddings[emotion] for emotion in emotion_labels])

# Compute cosine similarity between emotion classes
cosine_sim_matrix = cosine_similarity(embedding_matrix)

# Convert similarity matrix to a DataFrame for better visualization
df_cosine_sim = pd.DataFrame(cosine_sim_matrix, index=emotion_labels, columns=emotion_labels)

# Plot heatmap of cosine similarity
plt.figure(figsize=(10, 8))
sns.heatmap(df_cosine_sim, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

# Formatting the heatmap
plt.title("Cosine Similarity Between Emotion Classes")
plt.xlabel("Emotion")
plt.ylabel("Emotion")

# Display the plot
plt.show()
```

The cosine similarity heatmap illustrates the relationships between different emotion
classes based on their text embeddings. High similarity values (closer to 1) indicate strong
semantic overlap between emotions, while lower values suggest more distinct emotional
expressions. Emotions such as anger, shame, and guilt exhibit strong similarity, implying
shared linguistic patterns. Meanwhile, joy appears slightly more distinct from negative
emotions like sadness and disgust. This suggests that some emotions may be harder to
separate due to overlapping word usage, emphasizing the need for more refined feature
extraction or contextual modeling.

## 3. LLM Model Selection & Implementation

For this study, we selected three Large Language Models (LLMs)—Mistral-7B, LLaMA-2, and
Flan-T5—to perform emotion classification. These models were chosen based on their
architecture, performance in natural language understanding tasks, and suitability for fine-tuning
on emotion-related datasets.

### Model 1: Mistral 7B

Mistral 7B is a state-of-the-art decoder-only model that significantly outperforms previous models of similar size.

It was selected due to its:

1. Efficient fine-tuning capabilities: Supports Low-Rank Adaptation (LoRA), reducing GPU memory requirements and enabling fast training.
2. Strong generalization ability: Performs well across NLP tasks, including classification and reasoning.
3. Lower computational cost compared to larger models like LLaMA 2-13B.

This model is well-suited for fine-tuning in environments where GPU memory is constrained while still maintaining high classification accuracy.

### Model 2: LLaMA 2 (7B)

LLaMA 2 (7B), developed by Meta, is one of the leading open-source general-purpose LLMs, known for its strong contextual understanding and performance in text-based tasks. It was selected for:

1. State-of-the-art performance in classification tasks.
2. Fine-tuning efficiency with QLoRA (Quantized LoRA), reducing memory overhead.
3. Better text comprehension and reasoning capabilities compared to smaller models.

LLaMA 2 (7B) is the ideal choice when maximizing classification accuracy is the priority, even at the cost of slightly longer fine-tuning times.

### Model 3: Flan-T5 Large

Flan-T5 Large, an instruction-tuned variant of T5 (Text-to-Text Transfer Transformer), is optimized for NLP tasks, making it an excellent choice for emotion classification. Key advantages include:

1. Small model size (770M parameters), leading to faster fine-tuning and inference.
2. Pre-trained on diverse instruction-following tasks, enhancing zero-shot and few-shot classification performance.
3. Lower training and inference costs, making it an efficient choice for rapid deployment.

Flan-T5 is particularly effective when computational resources are limited, yet high-quality classification is required.

Zero-shot prompting directly asks the model to classify text without providing any prior examples.

### Example API Call for GPT-4 (Zero-Shot Approach):

```python
import openai

def classify_emotion(text):
    """
    Predicts the emotion of the given text using OpenAI's GPT-4 model.

    Args:
        text (str): The input text to analyze.

    Returns:
        str: The predicted emotion label from the predefined categories.
    """
    prompt = f"""
    Analyze the sentiment of the following text and classify it into one of the emotions:
    [joy, anger, fear, sadness, disgust, surprise].

    Text: "{text}"

    Response:
    """
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )

    return response["choices"][0]["message"]["content"].strip()

# Example usage
if __name__ == "__main__":
    sample_text = "I am so happy and excited today!"
    predicted_emotion = classify_emotion(sample_text)
    print(f"Classified Emotion: {predicted_emotion}")
```

### Few-Shot Prompting
Few-shot learning provides examples in the prompt, improving model accuracy.

Prompt Template:

### Example API Call for GPT-4 (Few-Shot Approach):

```python
import openai

def classify_emotion_few_shot(text):
    """
    Predicts the emotion of the given text using a few-shot learning approach with GPT-4.

    Args:
        text (str): The input text to analyze.

    Returns:
        str: The classified emotion label from the predefined categories.
    """
    prompt = f"""
    Classify the sentiment of the given text into one of the emotions:
    [joy, anger, fear, sadness, disgust, surprise].

    Below are examples for reference:

    1. "I just got a promotion at work!" → joy
    2. "I can't believe they betrayed me." → anger
    3. "I feel so anxious before my exam." → fear
    4. "Losing my pet was heartbreaking." → sadness
    5. "That food looks awful." → disgust
    6. "I was amazed by the magic trick!" → surprise

    Text: "{text}"

    Response:
    """
    response = openai.ChatCompletion.create(
        model="gpt-4",
        messages=[{"role": "user", "content": prompt}]
    )

    return response["choices"][0]["message"]["content"].strip()

# Example usage
if __name__ == "__main__":
    example_text = "I just got a new puppy!"
    predicted_emotion = classify_emotion_few_shot(example_text)
    print(f"Classified Emotion: {predicted_emotion}")
```

This approach does not require fine-tuning.
It generalizes well but might lack domain-specific adaptation.

### 3.2.3. Fine-Tuning Flan-T5 Large
Flan-T5 requires text-to-text format, so we need to modify the dataset slightly.

### Load Dataset & Tokenize

```python
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration, Trainer, TrainingArguments
from datasets import load_dataset

# Check for GPU or fallback to CPU
device = torch.device("cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu")
print(f"Using device: {device}")

# Load tokenizer for Flan-T5
MODEL_NAME = "google/flan-t5-small"
flan_tokenizer = T5Tokenizer.from_pretrained(MODEL_NAME)

# Load dataset from JSONL files
flan_dataset = load_dataset("json", data_files={"train": "train.jsonl", "test": "test.jsonl"})

# Preprocess dataset: format inputs and labels for Flan-T5
def tokenize_flan(examples):
    inputs = [f"Classify emotion: {text}" for text in examples["text"]]
    labels = [text for text in examples["label"]]

    model_inputs = flan_tokenizer(inputs, truncation=True, padding="max_length", max_length=128)
    with flan_tokenizer.as_target_tokenizer():
        model_inputs["labels"] = flan_tokenizer(labels, truncation=True, padding="max_length", max_length=10)["input_ids"]

    return model_inputs

# Apply tokenization to the dataset
flan_dataset = flan_dataset.map(tokenize_flan, batched=True)

# Convert dataset format to PyTorch tensors
flan_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

### Define Training Arguments & Model

```python
import torch
torch.cuda.empty_cache()
torch.cuda.ipc_collect()
print("CUDA Cache Cleared!")
```

```python
# Load Flan-T5 model and move to the appropriate device
flan_model = T5ForConditionalGeneration.from_pretrained(MODEL_NAME).to(device)

# Adjust batch size based on available hardware
batch_size = 2 if torch.cuda.is_available() else 1  # Use a larger batch size for GPU

# Define training parameters
training_args = TrainingArguments(
    output_dir="./flan_t5_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    logging_dir="./logs",
    load_best_model_at_end=True,
    save_total_limit=1,
    gradient_accumulation_steps=4 if torch.cuda.is_available() else 8,  # Adjust gradient accumulation
    fp16=torch.cuda.is_available(),  # Enable FP16 for GPU training
    dataloader_pin_memory=False  # Prevents memory pinning issues
)

# Initialize Trainer with model, dataset, and training parameters
trainer = Trainer(
    model=flan_model,
    args=training_args,
    train_dataset=flan_dataset["train"],
    eval_dataset=flan_dataset["test"]
)

# Start the training process
print(f"Starting training on: {device}")
trainer.train()
```

### Fine-Tuning LLaMA 2 (7B)
LLaMA 2-7B is a high-performance model, and we can fine-tune it using QLoRA for efficiency.

### Load Dataset & Tokenize

```python
import torch
from transformers import LlamaTokenizer, LlamaForSequenceClassification, Trainer, TrainingArguments, BitsAndBytesConfig
from datasets import load_dataset
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training

# Check for GPU availability
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Load tokenizer and set padding token
MODEL_NAME = "meta-llama/Llama-2-7b-hf"
llama_tokenizer = LlamaTokenizer.from_pretrained(MODEL_NAME)
llama_tokenizer.pad_token = llama_tokenizer.eos_token

# Load dataset
llama_dataset = load_dataset("json", data_files={"train": "train.jsonl", "test": "test.jsonl"})

# Define label mapping for emotions
emotion_labels = {"anger": 0, "disgust": 1, "fear": 2, "guilt": 3, "joy": 4, "sadness": 5, "shame": 6}

# Tokenization function
def tokenize_llama(examples):
    inputs = llama_tokenizer(examples["text"], truncation=True, padding="max_length", max_length=256)
    inputs["labels"] = [emotion_labels[label] for label in examples["label"]]
    return inputs

# Apply tokenization
llama_dataset = llama_dataset.map(tokenize_llama, batched=True)
llama_dataset.set_format(type="torch", columns=["input_ids", "attention_mask", "labels"])
```

### Define Training Arguments & Model

```python
import torch
torch.cuda.empty_cache()
torch.cuda.ipc_collect()
print("CUDA Cache Cleared!")
```

```python
# Enable 4-bit Quantization for Memory Efficiency
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

# Load model with quantization and distributed device mapping
llama_model = LlamaForSequenceClassification.from_pretrained(
    MODEL_NAME,
    num_labels=7,
    quantization_config=quantization_config,
    device_map="auto",
    torch_dtype=torch.float16,
    attn_implementation="sdpa"
)

# Prepare model for LoRA fine-tuning
llama_model = prepare_model_for_kbit_training(llama_model)

# Apply LoRA adapters for efficient fine-tuning
lora_config = LoraConfig(
    r=8,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj"],
    bias="none",
    task_type="SEQ_CLS"
)
llama_model = get_peft_model(llama_model, lora_config)

# Ensure caching is disabled for training
llama_model.config.use_cache = False

# Enable memory-efficient techniques
batch_size = 1
llama_model.gradient_checkpointing_enable()

# Define training arguments
training_args = TrainingArguments(
    output_dir="./llama2_results",
    evaluation_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=3,
    per_device_train_batch_size=batch_size,
    per_device_eval_batch_size=batch_size,
    logging_dir="./logs",
    load_best_model_at_end=True,
    save_total_limit=1,
    gradient_accumulation_steps=8,
    fp16=True if torch.cuda.is_available() else False,
    dataloader_pin_memory=False,
)

# Initialize Trainer
trainer = Trainer(
    model=llama_model,
    args=training_args,
    train_dataset=llama_dataset["train"],
    eval_dataset=llama_dataset["test"]
)

# Train the model
print("Starting training on:", device)
trainer.train()
```

### Step 3: Define Function to Evaluate Models

```python
def evaluate_model(model_path, model_name):
    """
    Evaluates a fine-tuned transformer model on the test dataset.

    Args:
        model_path (str): Path to the fine-tuned model directory.
        model_name (str): Name of the model for display purposes.

    Returns:
        str: Classification report summarizing model performance.
    """
    from transformers import AutoModelForSequenceClassification, AutoTokenizer
    from sklearn.metrics import classification_report
    import torch

    print(f"\nEvaluating {model_name}...\n")

    # Determine available device (GPU or CPU)
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # Load the fine-tuned model with optimized settings
    model = AutoModelForSequenceClassification.from_pretrained(
        model_path,
        num_labels=7,  # Ensure the model expects 7 emotion classes
        torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32,
        low_cpu_mem_usage=True,
        device_map="auto"
    ).to(device)

    # Load the tokenizer
    tokenizer = AutoTokenizer.from_pretrained(model_path)

    # Assign EOS token as pad token if missing
    if tokenizer.pad_token is None:
        print("Padding token missing! Assigning EOS token as pad token.")
        tokenizer.pad_token = tokenizer.eos_token

    # Perform evaluation (batch size = 1 to prevent memory issues)
    preds = []
    model.eval()

    with torch.no_grad():
        for text in test_texts:
            inputs = tokenizer(text, padding=True, truncation=True, max_length=256, return_tensors="pt")
            inputs = {key: val.to(device) for key, val in inputs.items()}

            outputs = model(**inputs)
            preds.append(torch.argmax(outputs.logits, dim=1).cpu().numpy()[0])

    # Generate and return classification report
    report = classification_report(true_labels, preds, target_names=labels_list)
    print(report)
    return report
```

```python
import torch
import os
from transformers import AutoModelForSequenceClassification, AutoTokenizer, BitsAndBytesConfig
from sklearn.metrics import classification_report
from datasets import load_dataset

# Set device (GPU or CPU)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Using device: {device}")

# Optimize memory usage
torch.cuda.empty_cache()
torch.backends.cudnn.benchmark = True  # Improves performance for repeated input sizes

# Enable 4-bit quantization for memory efficiency
quantization_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_compute_dtype=torch.float16,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_use_double_quant=True
)

# Model loading parameters
load_kwargs = {
    "device_map": "auto",
    "quantization_config": quantization_config,
    "torch_dtype": torch.float16,
    "offload_folder": "offload"
}

# Define model paths
model_paths = {
    "Mistral-7B": "/content/drive/My Drive/GenAI/Midterm/ISEAR/mistral_finetuned_full",
    "LLaMA-2": "/content/drive/My Drive/GenAI/Midterm/ISEAR/llama2_finetuned_full",
    "Flan-T5": "/content/drive/My Drive/GenAI/Midterm/ISEAR/flan_t5_finetuned_full",
}

# Load test dataset
test_file_path = "/content/drive/My Drive/GenAI/Midterm/ISEAR/Data/test.jsonl"
dataset = load_dataset("json", data_files={"test": test_file_path})["test"]

# Define emotion labels
labels_list = ["anger", "disgust", "fear", "guilt", "joy", "sadness", "shame"]
test_texts = dataset["text"]
true_labels = [labels_list.index(label) for label in dataset["label"]]

# Function to evaluate model performance
def evaluate_model(model_path, model_name):
    print(f"\nEvaluating {model_name}...\n")

    try:
        # Load model and tokenizer
        model = AutoModelForSequenceClassification.from_pretrained(model_path, num_labels=7, **load_kwargs)
        tokenizer = AutoTokenizer.from_pretrained(model_path)

        # Assign padding token if missing
        if tokenizer.pad_token is None:
            print("Padding token missing. Assigning EOS token as pad token.")
            tokenizer.pad_token = tokenizer.eos_token

        # Perform batch-wise inference
        batch_size = 1
        preds = []
        model.eval()
        with torch.no_grad():
            for i in range(0, len(test_texts), batch_size):
                batch_texts = test_texts[i : i + batch_size]
                inputs = tokenizer(batch_texts, padding=True, truncation=True, max_length=256, return_tensors="pt")
                inputs = {key: val.to(device) for key, val in inputs.items()}

                outputs = model(**inputs)
                preds.extend(torch.argmax(outputs.logits, dim=1).cpu().numpy())

        # Generate and save classification report
        report = classification_report(true_labels, preds, target_names=labels_list)
        print(report)

        report_dir = "/content/drive/My Drive/GenAI/Midterm/ISEAR/reports"
        os.makedirs(report_dir, exist_ok=True)
        report_path = os.path.join(report_dir, f"{model_name}_classification_report.txt")

        with open(report_path, "w") as f:
            f.write(report)

        print(f"Classification report saved at: {report_path}\n")
        return report

    except torch.cuda.OutOfMemoryError:
        print(f"Skipping {model_name} due to Out-of-Memory (OOM) error.")
        return f"{model_name} - OOM Error"

# Evaluate models sequentially
for model_name, model_path in model_paths.items():
    torch.cuda.empty_cache()
    evaluate_model(model_path, model_name)
```

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set up report directory
report_dir = "/content/drive/My Drive/GenAI/Midterm/ISEAR/reports"

# Define models and their respective classification report file paths
model_files = {
    "Mistral-7B": os.path.join(report_dir, "Mistral-7B_classification_report.txt"),
    "LLaMA-2": os.path.join(report_dir, "LLaMA-2_classification_report.txt"),
    "Flan-T5": os.path.join(report_dir, "Flan-T5_classification_report.txt"),
}

# Initialize data storage for extracted results
results = []

# Function to parse classification report and extract relevant metrics
def parse_classification_report(report_path, model_name):
    with open(report_path, "r") as file:
        lines = file.readlines()

    print(f"Processing Report: {report_path}")  # Debugging
    metrics = []

    # Adjusted regex pattern to support multi-word labels and handle dynamic spacing
    pattern = re.compile(r"([\w\s-]+)\s+([\d.]+)\s+([\d.]+)\s+([\d.]+)\s+(\d+)")

    for line in lines:
        match = pattern.match(line)
        if match:
            label, precision, recall, f1_score, support = match.groups()
            label = label.strip()  # Clean up extra spaces

            # Exclude overall accuracy and average metrics
            if "accuracy" in label.lower() or "avg" in label.lower():
                continue

            metrics.append({
                "Model": model_name,
                "Label": label,
                "Precision": float(precision),
                "Recall": float(recall),
                "F1-Score": float(f1_score),
                "Support": int(support)
            })
        else:
            print(f"Skipped line: {line.strip()}")  # Debugging

    return metrics

# Load and parse reports for each model
for model_name, report_path in model_files.items():
    if os.path.exists(report_path):
        report_metrics = parse_classification_report(report_path, model_name)

        if report_metrics:
            results.extend(report_metrics)
        else:
            print(f"No data extracted from {model_name} report. Check formatting!")

# Convert extracted results into a DataFrame
df_results = pd.DataFrame(results)

# Save extracted results as a CSV file for further analysis
if not df_results.empty:
    csv_path = os.path.join(report_dir, "model_comparison_results.csv")
    df_results.to_csv(csv_path, index=False)
    print(f"Model comparison results saved at: {csv_path}")
else:
    print("No valid data extracted. Check report formats!")

# Generate performance visualization if data is available
if not df_results.empty:
    plt.figure(figsize=(12, 6))
    sns.barplot(x="Label", y="F1-Score", hue="Model", data=df_results)
    plt.xticks(rotation=45)
    plt.title("Model Performance Comparison (F1-Score)")
    plt.ylabel("F1-Score")
    plt.xlabel("Emotion Labels")
    plt.legend(title="Models")
    plt.show()
else:
    print("No data available for visualization.")
```
